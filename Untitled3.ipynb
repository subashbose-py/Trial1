```python
# -- coding: utf-8 --
"""
Created on Wed Jan  5 22:26:37 2022

@author: Subash
"""

# -- coding: utf-8 --
"""
Created on Wed Jan  5 20:56:37 2022

@author: Subash
"""


from netmiko import ConnectHandler

from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
import smtplib
from email import encoders


iosv_l2_s1 = {
    'device_type': 'cisco_ios',
    'ip': '192.168.122.71',
    'username': 'subash',
    'password': 'cisco123'
}


iosv_l2_r1 = {
    'device_type': 'cisco_ios',
    'ip': '192.168.122.72',
    'username': 'subash',
    'password': 'cisco123'
}

all_devices = [iosv_l2_s1, iosv_l2_r1]

def sendMail(data):
    print("data is-  " + data)
    msg = MIMEMultipart()

    password = "XXXX"
    msg['From'] = "killgodzz@gmail.com"
    msg['To'] = "subash.boze@gmail.com"

    if "s" == data:
        file = "Switch_Config.txt"
        attachment = open(file,'rb')
    elif "r" == data:
        file = "Router_Config.txt"
        attachment = open(file,'rb')

    obj = MIMEBase('application','octet-stream')

    obj.set_payload((attachment).read())
    encoders.encode_base64(obj) 
    obj.add_header('Content-Disposition',"attachment; filename= "+file)
    msg.attach(obj)

    my_message = msg.as_string()

    email_session = smtplib.SMTP('smtp.gmail.com',587)
    email_session.starttls()
    email_session.login(msg['From'], password)

    email_session.sendmail(msg['From'], msg['To'], my_message)

    email_session.quit()

    print ("Mail have been sent for " + data + " , please check :)")



for devices in all_devices:
    host_name = " "
    net_connect = ConnectHandler(**devices)

    print ("Establishing Connection")

    host_name= net_connect.send_command("sh run | inc hostname")

    print(host_name)

    config_commands = net_connect.send_command("show log")
    
    if "S" in  host_name:
        switch_file = host_name + "\n"+ config_commands
        f = open("Switch_Config.txt", "a")
        f.write(switch_file)
        sendMail("s")
        f.close()
        host_name = " "
    elif "R" in host_name:
        router_file = host_name + "\n"+ config_commands
        f = open("Router_Config.txt", "a")
        f.write(router_file)
        sendMail("r")
        f.close()
        host_name = " "
```
